```python
# ER1, d=4
import logging
import numpy as np

from torch.utils.data import Dataset

from synthetic_dataset import SyntheticDataset
from synthetic_dataset import dataset_based_on_B


# Load dataset
n, d = 10000, 4
graph_type, degree = 'ER', 1    # ER1 graph
B_scale = 1.0
noise_type = 'gaussian_ev'

Dataset = SyntheticDataset(n, d, graph_type, degree,
                          noise_type, B_scale, seed=7)
data = Dataset.X
R_hat = np.cov(data.T)
print(R_hat.shape)
d = R_hat.shape[0]
Omega_true = np.eye(d)
Theta_hat = inv(R_hat)
# A_init = np.linalg.cholesky(Theta_hat)
# A_init = A_init + 0.01 * np.random.randn(*A_init.shape)
np.random.seed(54)
B_init = np.random.randn(*R_hat.shape)
B_true = Dataset.B
G_true = Dataset.B_bin
beta = 5
lam = 0.5



A_true = (np.eye(d) - B_true) @ inv(sqrtm(Omega_true))
print("G_true = \n",G_true)
likelihood_true = - 2 * np.log(np.linalg.det(A_true)) + np.trace(A_true.T @ R_hat @ A_true)
print("likelihood_true = ", likelihood_true)
true_penalty = lam * np.count_nonzero(A_true[~np.eye(A_true.shape[0], dtype=bool)])
f_true = likelihood_true + true_penalty

B_final,G_final, info = train_gs_reset(
    Rhat_np = R_hat,
    lam = lam,
    delta = 1e-6,
    tau_start = 1,
    tau_end = 1,
    B_init = B_init,
    max_steps = 5000,
    history_every=10,
    seed = 2,
    printreset= True
)
    
print("G_est = \n",G_final)


eps = 1e-6
abs_B = np.clip(np.abs(B_init), eps, 1 - eps)

# 计算 U_init = logit(|B_init|)
U_init = np.log(abs_B / (1 - abs_B))

# 计算 P_init = sign(B_init) * |B_init|
P_init = np.sign(B_init) * abs_B

# 打印结果
print("U_init =\n", np.round(U_init,2))
print("P_init =\n",np.round(B_init,2))

print("U_final = \n", np.array2string(info["U_final"], formatter={'float_kind':lambda x: f"{x:.2f}"}))
print("P_final = \n", np.array2string(info["P_final"], formatter={'float_kind':lambda x: f"{x:.2f}"}))
print("Final Loss = ", info["final_loss"])
print("Final penalty = ", info["final_penalty"])
print("Final likelihood = ", info["final_likelihood"])
print("likelihood_true = ", likelihood_true)
```

## Test Likelihood Convergence Ratio

```python
# ============================================================
# Experiment: convergence ratio of likelihood to true likelihood
# ============================================================

# ------------------ 1. 数据准备 ------------------
n, d = 10000, 4
graph_type, degree = 'ER', 1    # ER1 graph
B_scale = 1.0
noise_type = 'gaussian_ev'

Dataset = SyntheticDataset(n, d, graph_type, degree,
                           noise_type, B_scale, seed=7)
data = Dataset.X
R_hat = np.cov(data.T)
B_true = Dataset.B
G_true = Dataset.B_bin
Omega_true = np.eye(d)

# ------------------ 2. 真值计算 ------------------
A_true = (np.eye(d) - B_true) @ inv(sqrtm(Omega_true))
likelihood_true = -2 * np.log(np.linalg.det(A_true)) + np.trace(A_true.T @ R_hat @ A_true)
lam = 0.5
true_penalty = lam * np.count_nonzero(A_true[~np.eye(A_true.shape[0], dtype=bool)])
f_true = likelihood_true + true_penalty

print("=== Ground Truth ===")
print("G_true =\n", G_true)
print("likelihood_true =", likelihood_true)
print("true objective (f_true) =", f_true)

# ------------------ 3. 多次随机实验 ------------------
t = 100                     # 实验次数
tau_start, tau_end = 1.0, 1.0
max_steps = 300
tol = 0.005                 # 收敛阈值（1%以内认为收敛）

likelihood_diffs = []
success_count = 0

for seed in range(t):
    np.random.seed(seed)
    B_init = np.random.randn(*R_hat.shape)

    B_final, G_final, info = train_gs_reset(
        Rhat_np=R_hat,
        lam=lam,
        delta=1e-6,
        tau_start=tau_start,
        tau_end=tau_end,
        B_init=B_init,
        max_steps=max_steps,
        history_every=100,
        seed=seed,
        printreset=False
    )

    likelihood_est = info["final_likelihood"]
    diff_ratio = abs(likelihood_est - likelihood_true) / abs(likelihood_true)
    likelihood_diffs.append(diff_ratio)

    is_converged = diff_ratio < tol
    success_count += int(is_converged)

    # status = "✅" if is_converged else "❌"
    # print(f"[{seed:02d}] likelihood_est={likelihood_est:.4f}, "
    #      f"diff_ratio={diff_ratio:.4f} -> {status}")

# ------------------ 4. 结果统计 ------------------
likelihood_diffs = np.array(likelihood_diffs)
convergence_ratio = success_count / t

print("\n=== Summary ===")
print(f"Total runs: {t}")
print(f"Tolerance (relative error): {tol:.2f}")
print(f"Converged runs: {success_count}")
print(f"Convergence ratio: {convergence_ratio:.3f}")
print(f"Average diff ratio: {np.mean(likelihood_diffs):.4f}")
print(f"Median diff ratio: {np.median(likelihood_diffs):.4f}")
```

```python
# ------------------ 3. 多次随机实验 ------------------
t = 100                     # 实验次数
tau_start, tau_end = 1.0, 1.0
max_steps = 10000
tol = 0.005                 # 收敛阈值（5%以内认为收敛）

likelihood_diffs = []
success_count = 0
penalty_count = 0

for seed in range(t):
    np.random.seed(seed)
    B_init = np.random.randn(*R_hat.shape)

    B_final, G_final, info = train_gs_reset(
        Rhat_np=R_hat,
        lam=lam,
        delta=1e-6,
        tau_start=tau_start,
        tau_end=tau_end,
        B_init=B_init,
        max_steps=max_steps,
        history_every=200,
        seed=seed,
        printreset=False
    )

    likelihood_est = info["final_likelihood"]
    penalty_est = info["final_penalty"]
    diff_ratio = abs(likelihood_est - likelihood_true) / abs(likelihood_true)
    likelihood_diffs.append(diff_ratio)

    is_converged = diff_ratio < tol
    success_count += int(is_converged)
    penalty_count += penalty_est

    status = "✅" if is_converged else "❌"
    print(f"[{seed:02d}] likelihood_est={likelihood_est:.4f}, "
          f"diff_ratio={diff_ratio:.4f} -> {status}")
    print(f"penalty = {penalty_est:.4f}")

# ------------------ 4. 结果统计 ------------------
likelihood_diffs = np.array(likelihood_diffs)
convergence_ratio = success_count / t
penalty_avg = penalty_count / t

print("\n=== Summary ===")
print(f"Total runs: {t}")
print(f"Tolerance (relative error): {tol:.2f}")
print(f"Converged runs: {success_count}")
print(f"Convergence ratio: {convergence_ratio:.3f}")
print(f"Average diff ratio: {np.mean(likelihood_diffs):.4f}")
print(f"Median diff ratio: {np.median(likelihood_diffs):.4f}")
print("Average penalty:",penalty_avg)
```
